```python
# Find the mode 
df_superstore.State.mode()
```

```python
# Find the percentage of null values
df_superstore.isnull().sum()/len(df_superstore)*100
```

# Segment

```python
# Find the percentage of null value in that column
# Find the percentage of null values
df_superstore.isnull().sum()/len(df_superstore)*100
```

```python
# This is the cat column so we can replace the null values with mode
df_superstore.Segment.mode()[0]
```

```python
# Impute the null values
df_superstore.Segment = df_superstore.Segment.fillna(df_superstore.Segment.mode()[0])
```

```python
# Find the percentage of null values
df_superstore.isnull().sum()/len(df_superstore)*100
```

```python
# Impute the null values
df_superstore.Region = df_superstore.Region.fillna(df_superstore.Region.mode()[0])
# Impute the null values
df_superstore["Ship Mode"] = df_superstore["Ship Mode"].fillna(df_superstore["Ship Mode"].mode()[0])
```

```python
# Find the percentage of null values
df_superstore.isnull().sum()/len(df_superstore)*100
```

# Sales     

```python
# Find the mean of that column
df_superstore.Sales.mean()
```

```python
# Fill the na values with the mean of that same column.
df_superstore.Sales.fillna(df_superstore.Sales.mean(),inplace=True)
```

```python
# Check the actions
df_superstore.isnull().sum()/len(df_superstore)*100
```

```python
# replace the null of Descount and profit with there respective mean
# Discount
df_superstore.Discount.fillna(df_superstore.Discount.mean(),inplace=True)

# Profit
df_superstore.Profit.fillna(df_superstore.Profit.mean(),inplace=True)
```

```python
# Check the actions
df_superstore.isnull().sum()/len(df_superstore)*100
```
